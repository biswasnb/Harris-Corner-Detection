## Harris corner
Implement the Harris corner detector to find and visualize corners in a grayscale image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image_path = "/content/img a1.jpg"  # Replace with your image path
img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Convert to float32 (required for Harris corner detection)
gray = np.float32(gray)

# Apply Harris Corner Detection
dst = cv2.cornerHarris(gray, blockSize=3, ksize=3, k=0.04)

# Dilate corner points to make them more visible
dst = cv2.dilate(dst, None)

# Threshold for marking corners
img[dst > 0.01 * dst.max()] = [0, 255, 0]  # Mark corners in green

# Display Original Image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

# Display Image with Detected Corners
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Harris Corner Detection")
plt.axis("off")

plt.show()
```
